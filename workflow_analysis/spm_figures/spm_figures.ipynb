```python
import numpy as np
import matplotlib.pyplot as plt

# Labels and Ranks
labels = ['indiv \n+ merge', 'merge \n+ mutation', 'merge \n+ freq', 'sift \n+ mutation', 'sift \n+ freq']
ranks = ['R 1', 'R 2', 'R 3', 'R 4', 'R 5', 'R 6', 'R 7', 'R 8', 'R 9']

# Bar plot data: SPM Time Deviation Percentage to Runtime Rank
data = [[0, 0.000738462, 4.067446153220394, 0, 4.0667076916820415],
    [0, 0.005907692, 0.005907692, 0, 0],
    [0, -0.111938461, 0, -0.038461538, -4.066707692],
    [0, 0.10529230752226298, -4.073353846, 0.059784615, 0],
    [-58.83559619, 11.045925595069859, -9.235553846, -0.059784615, 0],
    [141.87555320894222, 0, -19.47739486, 0, -19.13438866],
    [-187.0996191, -10.5945856, 11.057733736760342, -0.91817358, 11.064432413682391],
    [52.299953945800006, -0.451339994, 0, -0.459086788, -1.778212384],
    [-7.075888014, 0, -11.05773374, 1.3772603674804216, -9.286220029],
]

# Line data: percentage contribution to total workflow time
workflow_pct = [
                0.537202677, 0.0240108, 0.149440893, 0.024822454, 0.150252547

                
                ]
workflow_pct_percent = [x * 100 for x in workflow_pct]  # Convert to percent for overlay

# Colors
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22']

# Plot setup
x = np.arange(len(labels)) * 1.2 # label positions
bar_width = 0.1

fig, ax1 = plt.subplots(figsize=(12, 5))

# Bar plots on primary y-axis
for i in range(len(ranks)):
    bars = ax1.bar(x + i * bar_width, data[i], width=bar_width, color=colors[i], label=ranks[i], zorder=3)
    
    # Add data labels for values exceeding +/-100%
    for j, val in enumerate(data[i]):
        xpos = x[j] + i * bar_width
        if abs(val) > 100:
            ax1.text(
                xpos,
                5 + (3 if val > 0 else -8),
                # f'{val:.1f}%',
                f'{int(val)}%',
                ha='center',
                va='bottom' if val > 0 else 'top',
                fontsize=12,
                rotation=90,
                zorder=6,
                color='black',
                bbox=dict(facecolor='white', edgecolor='none', alpha=0.7, boxstyle='round,pad=0.2')
            )

# Add tick markers at zero values
for i in range(len(ranks)):
    for j, val in enumerate(data[i]):
        xpos = x[j] + i * bar_width
        ax1.plot(xpos, 0, marker='|', color='black', markersize=10, zorder=4)

# Axis setup for primary y-axis
centered_x = x + ((len(ranks) - 1) * bar_width) / 2
ax1.set_xticks(centered_x)
ax1.set_xticklabels(labels, fontsize=18)

# Y-axis ticks and dashed lines every 25% from -100 to 100
y_ticks = np.arange(-100, 120, 20)
ax1.set_yticks(y_ticks)
ax1.set_ylim(-100, 100)
ax1.set_ylabel('SPM Deviation Time (%)', fontsize=18)
ax1.set_xlabel('1000 Genomes Producer-Consumer Pairs', fontsize=18)

# Add dashed lines at each tick
for y in y_ticks:
    ax1.axhline(y, color='gray', linestyle='--', linewidth=0.5, zorder=0)

# Optional: remove default grid to avoid double lines
ax1.grid(False)
    
# Secondary y-axis for workflow percent
ax2 = ax1.twinx()
ax2.plot(centered_x, workflow_pct_percent, marker='o', linestyle='--', color='black',
         label='P-C % of Total Run Time', linewidth=2, zorder=5)
ax2.set_ylim(-100, 100)
ax2.tick_params(axis='y', labelsize=14)
# ax2.tick_params(axis='y', which='both', left=False, right=False, labelleft=False)

# Combine legends
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, title='Rank', fontsize=16,title_fontsize=14,
           ncol=5, loc='lower center', bbox_to_anchor=(0.58, 0.0), columnspacing=1)

# Final layout and save
plt.tight_layout()
plt.savefig("1kg_spm_time_deviation.pdf", format='pdf', bbox_inches='tight')
plt.show()
```
